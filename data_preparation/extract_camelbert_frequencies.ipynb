### What to do now: sort the df, count until reaching the desired frequency thresholds (from Train dataset on Samer Corpus)

```python
big_sum = sum(cleaned[1])
```

```python
l3_threshold = 0.8652
l4_threshold = 0.0900
l5_threshold = 0.0448

words_and_levels = {}

```

```python
cleaned = cleaned.sort_values(1, ascending = False)
```

```python
cleaned
```

```python
sum = 0

for row in tqdm(cleaned.iterrows()):
  sum += int(row[1][1])
  if sum > big_sum * (l3_threshold + l4_threshold):
    words_and_levels[row[1][0]] = 5
  elif sum > big_sum * l3_threshold:
    words_and_levels[row[1][0]] = 4
  else:
    words_and_levels[row[1][0]] = 3
```
